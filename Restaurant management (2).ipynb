**Online RMS**

```python
def get_user_choice(options):
    for i, option in enumerate(options, start=1):
        print(f"{i}. {option}")

    user_input = int(input("Enter the number of your choice: "))
    while user_input < 1 or user_input > len(options):
        user_input = int(input("Invalid choice. Please enter a valid number: "))

    return options[user_input - 1]


class FoodItem:
    def __init__(self, name, rating, preparation_time):
        self.name = name
        self.rating = rating
        self.preparation_time = preparation_time

def apply_discount(total):
    # Example: Apply a 10% discount during a festival
    discount_percentage = 0.10
    discounted_total = total - (total * discount_percentage)
    return discounted_total

def calculate_shipping_charges(total_cost):
    # Example: Apply shipping charges if total cost is less than $150
    if total_cost < 150:
        return 10  # Example shipping charge
    else:
        return 0  # No shipping charge if total cost is $150 or more
    
def calculate_delivery_time(selected_food, selected_combo):
    food_preparation_time = selected_food.preparation_time
    combo_preparation_time = sum(item.preparation_time for item in selected_combo["items"])
    
    total_preparation_time = max(food_preparation_time, combo_preparation_time)
    
    # Add additional time for delivery, for example, 15 minutes
    delivery_time = 15
    
    estimated_delivery_time = total_preparation_time + delivery_time
    return estimated_delivery_time

def get_delivery_contact(total_cost):
    shipping_charges = calculate_shipping_charges(total_cost)
    return f"Contact us at: 123-456-7890\nShipping Charges: ${shipping_charges}"

def get_address():
    # Example: Prompt user to input their address
    return input("Please enter your delivery address: ")

def get_payment_method():
    # Example: Allow user to select payment method
    payment_methods = ["Credit Card", "Debit Card", "Cash on Delivery"]
    return get_user_choice(payment_methods)

def choose_food_delivery_app():
    # Example: Allow user to select a food delivery app
    food_delivery_apps = ["Zomato", "Swiggy", "Uber Eats", "Foodpanda"]
    for i, app in enumerate(food_delivery_apps, start=1):
        print(f"{i}. {app}")

    user_input = int(input("Enter the number of your choice: "))
    while user_input < 1 or user_input > len(food_delivery_apps):
        user_input = int(input("Invalid choice. Please enter a valid number: "))

    return food_delivery_apps[user_input - 1]

# Choose food delivery app
chosen_app = choose_food_delivery_app()
print(f"You have chosen {chosen_app} for food delivery. Enjoy your meal!")


def search_food(items, query):
    matching_items = [item for item in items if query.lower() in item.name.lower()]
    return matching_items

food_items = [
    FoodItem("Paneer Tikka", 4.5, 20),
    FoodItem("Veg Biryani", 4.0, 25),
    FoodItem("Dal Makhani", 4.2, 15),
    FoodItem("Chicken Tikka", 4.8, 30),
    FoodItem("Butter Chicken", 4.7, 35),
    FoodItem("Fish Curry", 4.6, 25),
]

# Search for food
search_query = input("Enter the food you want to order: ")
matching_food = search_food(food_items, search_query)

if matching_food:
    print("Matching food items:")
    for i, item in enumerate(matching_food, start=1):
        print(f"{i}. {item.name} (Rating: {item.rating})")
    selected_food = get_user_choice(matching_food)
    print(f"You have selected {selected_food.name} (Rating: {selected_food.rating}).")

    veg_or_nonveg = get_user_choice(["Vegetarian", "Non-vegetarian"])

    # Vegetarian Menu
    if veg_or_nonveg == "Vegetarian":
        veg_starters = [FoodItem("Paneer Tikka", 4.5,20), FoodItem("Veg Manchurian", 4.2,20), FoodItem("Spring Rolls", 4.2,15)]
        veg_starter = get_user_choice(veg_starters)
        print(f"You have selected {veg_starter.name} as your vegetarian starter. (Rating: {veg_starter.rating})")

        veg_mains = [FoodItem("Dal Makhani", 4.2,15), FoodItem("Jeera Aloo", 4.0,30), FoodItem("Kadai Paneer", 4.5,25)]
        veg_main = get_user_choice(veg_mains)
        print(f"You have selected {veg_main.name} as your vegetarian main course. (Rating: {veg_main.rating})")

        veg_combos = [
            {"combo": "Veg Biryani Combo", "items": [FoodItem("Veg Biryani", 4.0,25), FoodItem("Raita", 4.0,10), FoodItem("Bread halwa", 4.2,10)]},
            {"combo": "Full meals", "items": [FoodItem("White rice", 4.0,10), FoodItem("Sambar", 4.2,20), FoodItem("Rasam", 4.5,15), FoodItem("Cord", 4.0,20), FoodItem("Papad", 4.0,10), FoodItem("Gulab jamun", 4.6,20)]},
            {"combo": "Chow-Chow Bath", "items": [FoodItem("Khara Bath", 4.2,15), FoodItem("Kesari Bath", 4.5,15), FoodItem("Chutney", 4.0,10)]},
            {"combo": "Dont want", "items": [FoodItem("Thank you", 5.0,0)]},
        ]
        veg_combo = get_user_choice([combo["combo"] for combo in veg_combos])
        selected_combo = next(combo for combo in veg_combos if combo["combo"] == veg_combo)
        print(f"{selected_combo['combo']}:")
        for item in selected_combo["items"]:
            print(f"\t- {item.name} (Rating: {item.rating})")

    # Non-vegetarian Menu
    else:
        nonveg_starters = [FoodItem("Chicken Tikka", 4.8,30), FoodItem("Mutton Seekh Kebab", 4.5,15), FoodItem("Fish Fingers", 4.2,15)]
        nonveg_starter = get_user_choice(nonveg_starters)
        print(f"{nonveg_starter.name} (Rating: {nonveg_starter.rating})")

        nonveg_mains = [FoodItem("Butter Chicken", 4.7,35), FoodItem("Mutton Rogan Josh", 4.5,15), FoodItem("Fish Curry", 4.6,25)]
        nonveg_main = get_user_choice(nonveg_mains)
        print(f"You have selected {nonveg_main.name} as your non-vegetarian main course. (Rating: {nonveg_main.rating})")

        nonveg_combos = [
            {"combo": "Non-veg full meals", "items": [FoodItem("Butter Chicken", 4.7,35), FoodItem("Mutton Rogan Josh", 4.5,15), FoodItem("Fish Curry", 4.6,25), FoodItem("Rice", 4.0,10), FoodItem("Roti", 4.0,10)]},
            {"combo": "Mughlai Combo", "items": [FoodItem("Chicken Tikka", 4.8,30), FoodItem("Mutton Seekh Kebab", 4.5,15), FoodItem("Fish Fingers", 4.2,15), FoodItem("Biryani", 4.0,20), FoodItem("Raita", 4.0,10)]},
            {"combo": "Dont want", "items": [FoodItem("Thank you", 5.0,0)]},
        ]
        nonveg_combo = get_user_choice([combo["combo"] for combo in nonveg_combos])
        selected_combo = next(combo for combo in nonveg_combos if combo["combo"] == nonveg_combo)
        print(f"You have selected the {selected_combo['combo']} combo, which includes:")
        for item in selected_combo["items"]:
            print(f"\t- {item.name} (Rating: {item.rating})")
else:
    print("No matching food items found. Please try again.")
    
# Display selected food item with preparation time
print(f"You have selected {selected_food.name} (Rating: {selected_food.rating}).")
print(f"Preparation time: {selected_food.preparation_time} minutes.")

# Apply discount if it's festival time
apply_discount_option = get_user_choice(["Apply discount?", "No, thanks"])
if apply_discount_option == "Apply discount?":
    total_cost = 100  # Example total cost
    discounted_total = apply_discount(total_cost)
    print(f"Total cost after discount: ${discounted_total}")    
    
    # Display delivery contact with shipping charges
    print(get_delivery_contact(discounted_total))
else:
    # Display delivery contact with shipping charges
    print(get_delivery_contact(total_cost))
    
# Display selected combo items with preparation times
for item in selected_combo["items"]:
    print(f"\t- {item.name} (Rating: {item.rating})")
    print(f"\t  Preparation time: {item.preparation_time} minutes.")
    
# Calculate and display estimated delivery time
estimated_delivery_time = calculate_delivery_time(selected_food, selected_combo)
print(f"Estimated delivery time: {estimated_delivery_time} minutes")

# Prompt user for delivery address
delivery_address = get_address()
print(f"Your delivery address: {delivery_address}")

# Select payment method
payment_method = get_payment_method()
print(f"Payment method chosen: {payment_method}")
```

**Hall Reservation**

```python
class Reservation:
    def __init__(self):
        self.date = None
        self.occasion = None
        self.theme = None
        self.hall_type_options = {
            '1': 'Party Hall',
            '2': 'Dining Hall',
            '3': 'Banquet Halls',
            '4': 'Private Dining Rooms',
            '5': 'Terraces or Outdoor Spaces',
            '6': 'Meeting Rooms',
            '7': 'VIP Sections'
        }
        self.hall_type = None
        self.time_of_day_options = {
            '1': 'Morning',
            '2': 'Afternoon',
            '3': 'Evening',
            '4': 'Night'
        }
        self.time_of_day_choice = None
        self.time_options = None
        self.time = None
        self.games = []
        self.number_of_people = None
        self.cost = 100
        self.payment_mode = None

    def set_date(self, date):
        self.date = date

    def set_occasion(self, occasion):
        self.occasion = occasion

    def set_theme(self, theme):
        self.theme = theme

    def hall(self, hall_type_choice):
        if hall_type_choice in self.hall_type_options:
            self.hall_type = self.hall_type_options[hall_type_choice]
            print(f'Selected hall type: {self.hall_type}')
            self.timing()
        else:
            print('Incorrect Choice')

    def timing(self):
        print("Choose your time of day:")
        print(self.time_of_day_options)
        self.time_of_day_choice = input("Select your time of day:")

        # Choose the appropriate time_options based on time_of_day_choice
        time_options = {
            '1': {
                '1': '9:00-11:00',
                '2': '12:00-2:00',
                '3': '3:00-5:00',
                '4': '7:00-9:00',
                '5': '10:00-12:00'
            },
            '2': {
                '1': '10:00-12:00',
                '2': '2:00-4:00',
                '3': '5:00-7:00',
                '4': '8:00-10:00',
                '5': '11:00-1:00'
            },
            '3': {
                '1': '10:30-12:30',
                '2': '1:30-3:30',
                '3': '4:30-6:30',
                '4': '7:30-9:30',
                '5': '10:30-2:00'
            },
            '4': {
                '1': '9:30-11:30',
                '2': '12:30-2:30',
                '3': '3:30-5:30',
                '4': '7:30-9:30',
                '5': '10:30-2:00'
            }
        }

        if self.time_of_day_choice in time_options:
            self.time_options = time_options[self.time_of_day_choice]
            print("Choose your time:")
            print(self.time_options)
            t = input("Select your time:")
            self.time = self.time_options.get(t, "Invalid time option")
            print(f'Booking is noted at {self.time}')
        else:
            print("Invalid time of day choice.")

    def set_games(self, games):
        self.games = games.split(',')

    def set_number_of_people(self, number_of_people):
        self.number_of_people = int(number_of_people)

    def calculate_cost(self):
        # Customize your cost calculation logic based on your requirements
        base_cost = 500
        games_cost = len(self.games) * 50
        # Additional cost based on the number of people
        people_cost = self.number_of_people * 10
        self.cost = base_cost + games_cost + people_cost

    def set_payment_mode(self, payment_mode):
        self.payment_mode = payment_mode

    def display_booking_details(self):
        print("Booking Details:")
        print(f"Date: {self.date}")
        print(f"Occasion: {self.occasion}")
        print(f"Theme: {self.theme}")
        print(f"Hall Type: {self.hall_type}")
        print(f"Time: {self.time_of_day_options[self.time_of_day_choice]}, {self.time}")
        print(f"Games: {', '.join(self.games)}")
        print(f"Number of People: {self.number_of_people}")
        print(f"Total Cost: ${self.cost}")
        print(f"Payment Mode: {self.payment_mode}")

def get_user_choice(options):
    for i, option in enumerate(options):
        print(f"{i+1}. {option}")

    user_input = int(input("Enter the number of your choice: "))
    while user_input < 1 or user_input > len(options):
        user_input = int(input("Invalid choice. Please enter a valid number: "))

    return options[user_input-1]

reservation = Reservation()

reservation.set_date(input("Enter the date:"))
reservation.set_occasion(input("Enter the occasion:"))
reservation.set_theme(input("Enter the theme:"))
reservation.hall(input("Enter your choice:"))
reservation.set_games(input("Enter the games (comma-separated):"))
reservation.set_number_of_people(input("Enter the number of people:"))

# Calculate total cost
reservation.calculate_cost()
print(f"Total Cost: ${reservation.cost}")

# Set payment mode
payment_mode = input("Enter payment mode (e.g., Credit Card, Cash, UPI): ")
reservation.set_payment_mode(payment_mode)

# Display booking details
reservation.display_booking_details()

def get_user_choice(options):
    for i, option in enumerate(options):
        print(f"{i+1}. {option}")

    user_input = int(input("Enter the number of your choice: "))
    while user_input < 1 or user_input > len(options):
        user_input = int(input("Invalid choice. Please enter a valid number: "))

    return options[user_input-1]

veg_or_nonveg = get_user_choice(["Vegetarian", "Non-vegetarian"])
if veg_or_nonveg == "Vegetarian":
    veg_starters = ["Paneer Tikka", "Veg Manchurian", "Spring Rolls"]
    veg_starter = get_user_choice(veg_starters)
    print(f"You have selected {veg_starter} as your vegetarian starter.")

    veg_mains = ["Dal Makhani", "Jeera Aloo", "Kadai Paneer"]
    veg_main = get_user_choice(veg_mains)
    print(f"You have selected {veg_main} as your vegetarian main course.")

    veg_combos = [
        {"combo": "Veg Biryani Combo", "items": ["Veg Biryani", "Raita", "Bread halwa"]},
        {"combo": "Full meals", "items": ["White rice", "Sambar", "Rasam", "Cord", "Papad", "Gulab jamun"]},
        {"combo": "Chow-Chow Bath", "items": ["Khara Bath", "Kesari Bath", "Chutney"]},
        {"combo": "Dont want", "items": ["Thank you"]},
    ]
    veg_combo = get_user_choice([combo["combo"] for combo in veg_combos])
    selected_combo = next(combo for combo in veg_combos if combo["combo"] == veg_combo)
    print(f"{selected_combo['combo']}:")
    for item in selected_combo["items"]:
        print(f"\t- {item}")

else:
    nonveg_starters = ["Chicken Tikka", "Mutton Seekh Kebab", "Fish Fingers"]
    nonveg_starter = get_user_choice(nonveg_starters)
    print(f"{nonveg_starter} ")

    nonveg_mains = ["Butter Chicken", "Mutton Rogan Josh", "Fish Curry"]
    nonveg_main = get_user_choice(nonveg_mains)
    print(f"You have selected {nonveg_main} as your non-vegetarian main course.")

    nonveg_combos = [
        {"combo": "Non-veg full meals", "items": ["Butter Chicken", "Mutton Rogan Josh", "Fish Curry", "Rice", "Roti"]},
        {"combo": "Mughlai Combo", "items": ["Chicken Tikka", "Mutton Seekh Kebab", "Fish Fingers", "Biryani", "Raita"]},
        {"combo": "Dont want", "items": ["Thank you"]},
    ]
    nonveg_combo = get_user_choice([combo["combo"] for combo in nonveg_combos])
    selected_combo = next(combo for combo in nonveg_combos if combo["combo"] == nonveg_combo)
    print(f"You have selected the {selected_combo['combo']} combo, which includes:")
    for item in selected_combo["items"]:
        print(f"\t- {item}")
```
